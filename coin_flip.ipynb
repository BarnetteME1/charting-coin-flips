```python
def new_coin_flip():
    # heads = 1 tails = 0
    return choice(["1", "0"])

def one_hundred_flips():
    hundred = []
    for _ in range(100):
        hundred.append(new_coin_flip())
    return hundred
```

```python
def hundred_thousand():
    large_number = []
    for _ in range(100000):
        large_number.append(one_hundred_flips())
    return large_number
```

```python
trials = hundred_thousand()
```

```python
heads_ratio = []
for trial in trials:
    heads_ratio.append(trial.count("1")/len(trial))
print(heads_ratio)
```
